# ReviewLens — Exploration

Quick walkthrough of the baseline pipeline on the bundled sample data.
Run `pip install -e .` (and `requirements-core.txt`) first.

```python
from reviewlens.pipeline import run_pipeline
from reviewlens.aggregate.summary import aspect_distribution, top_loved, top_hated

result = run_pipeline()  # defaults to data/sample/sample_reviews.csv
print(len(result.reviews), 'reviews ->', len(result.aspects), 'aspect mentions')
result.aspects.head()
```

```python
# Per-theme sentiment distribution
aspect_distribution(result.aspects, group_col='theme')
```

```python
print('LOVED:')
display(top_loved(result.aspects, group_col='theme'))
print('HATED:')
display(top_hated(result.aspects, group_col='theme'))
```

```python
# Where ABSA beats a single score: reviews with mixed per-aspect sentiment
mixed = result.aspects.groupby('review_id')['aspect_sentiment'].nunique()
result.aspects[result.aspects['review_id'].isin(mixed[mixed > 1].index)]
```
